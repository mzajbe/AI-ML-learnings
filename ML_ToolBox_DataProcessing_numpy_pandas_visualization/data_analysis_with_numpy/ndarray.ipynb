```python
arr = [1,2,3]

print(arr * 5) #its multiply the whole array 5 times

arr = [x*5 for x in arr]

print(arr)
```

```python

import numpy as np


arr = [1,2,3]
arr1 = np.array(arr)  # using existing arr variable

arr1 = arr1 * 5

print(arr1)
```

Ndarray

```python
import numpy as np

#1d array
arr1 = np.array([1,2,3,4,5])

print(type(arr1))
print(arr1.ndim)


#2d array
arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2.ndim)


#3d array
arr3 = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print(arr3.ndim)
```
